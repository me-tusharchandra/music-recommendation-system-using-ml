# Exploratory Data Analysis, Visualization and Feature Engineering

```python
# Typical Length of an album

track_numbers = df['track_number'].unique()
old_count = None
        
albums = df.groupby(['album_id','album'])['album_id'].count()
albums = albums.reset_index(name='Album Length')
album_lengths = albums.groupby('Album Length')['Album Length'].count()
album_lengths = album_lengths.reset_index(name='Count')
album_lengths = album_lengths.head(20)
album_lengths.plot('Album Length', 'Count', kind='bar', legend=None, ylabel='Count');
album_lengths.sort_values(by='Count', ascending=False, inplace=True)
album_lengths_tab = album_lengths.to_string(index=False)

df = df.merge(albums[['album_id','Album Length']], 
                    how='inner', on='album_id')
```

```python
# Convert the boolean values into integers
df['explicit'] = df['explicit'].astype(int)
df
```

> Focusing on ten-track albums, we can use boxplots to explore the attributes of songs at different track numbers. (Some attributes are not plotted because their boxplots are not very informative.)

> These plots suggest that songs tend to decrease in energy from the beginning to the end of an album, while increasing in acousticness. Relatively long songs may be more common at the end of an album as well.

```python
songs_10track = df[df['Album Length']==10]

attributes = list(df.columns[9:21])
atts_by_track = []

for att in attributes:
    att_by_track = [songs_10track[
                    songs_10track['track_number']==track_number][att]
                    for track_number in range(1,11)]
    atts_by_track.append(att_by_track)

skipped_atts = ['key','loudness','mode','liveness','speechiness']
plt_height_factor = 3.5
fig, ax = plt.subplots(len(attributes) - len(skipped_atts), 1)
fig.set_size_inches(12,
                    plt_height_factor*(len(attributes)-len(skipped_atts)))
counter = 0
for a in range(0, len(attributes)):
    if attributes[a] not in skipped_atts:
        ax[counter].boxplot(atts_by_track[a])
        ax[counter].set_title(attributes[a].title())
        counter += 1
plt.show()
```

> Plotting the trends of songs for these attributes from 1960 on. We include mode as an attribute to check for trends in how many songs are in a major key as opposed to a minor key. We can see that overall, songs increased in danceability and energy, while decreasing in acousticness. Also, use of minor keys became more common. There was also a reversal in the overall trend during the 80s.

```python
import numpy as np
from scipy.interpolate import CubicSpline

songs_year = df[df['year'] >= 1960]
songs_year = songs_year[['year'] + attributes]
songs_year = songs_year.groupby('year').mean()
songs_year['year'] = songs_year.index

songs_year['decade'] = songs_year.index // 10 % 10 * 10
songs_year['decade'] = songs_year.decade.apply(lambda d: str(d).zfill(2))
songs_decade = songs_year.groupby('decade').mean()
songs_decade = songs_decade.sort_values(by='year')

year = list(songs_year.index)
decade = list(songs_decade.year)
#print(songs_decade)

fig, ax = plt.subplots(4, 2)
fig.set_size_inches(12, 16)
counter = 0

if 'mode' in skipped_atts:
    skipped_atts.remove('mode')
    
for att in attributes:
    if att in skipped_atts:
        continue
    row = counter // 2
    col = counter % 2
    counter += 1
    
    data = list(songs_year[att])
    decade_data = list(songs_decade[att])
    #tried polynomial fitting, but fit is not great
    #quad_fit = np.poly1d(np.polyfit(year, data, 5))
    lin = np.linspace(1960, 2020, 50)
    cs = CubicSpline(decade, songs_decade[att])
    
    ax[row,col].set_title(att.title())
    ax[row,col].scatter(year, data)
    ax[row,col].scatter(decade, decade_data, color='green')
    #plt.plot(quad_lin, quad_fit(quad_lin), color='green')
    ax[row,col].plot(lin, cs(lin), color='green')
    
plt.show();
```

```python
# Except id, dropping all the other non numeric columns
newDf = df.drop(columns=['artists','album','name','album_id', 'artist_ids', 'track_number', 'Album Length'])
newDf
```

```python
import matplotlib.pyplot as plt
import math

# New set of variables
variables = ['disc_number', 'danceability', 'explicit', 'energy', 'key',
            'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
            'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature', 'year']

# Calculate the number of rows and columns for the grid layout
num_variables = len(variables)
num_rows = math.ceil(math.sqrt(num_variables))
num_cols = math.ceil(num_variables / num_rows)

# Set the number of bins for the histograms
num_bins = 20

# Same color palette provided in the original code
colors = ['#FF5733', '#FFC300', '#C70039', '#900C3F', '#581845', '#FF7F50', '#FFD700', '#FF1493',
          '#FF00FF', '#00FF00', '#0000FF', '#FF4500', '#00FFFF', '#8A2BE2', '#FF69B4', '#00BFFF',
          '#7FFF00', '#FF8C00', '#8B0000', '#9400D3']

# Create the grid of subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(20, 20))

# Iterate over each variable and create a histogram in the corresponding subplot
for i, variable in enumerate(variables):
    row = i // num_cols
    col = i % num_cols

    # Create the histogram in the corresponding subplot
    axs[row, col].hist(newDf[variable], bins=num_bins, color=colors[i % len(colors)], edgecolor='black')

    # Set the labels and title for the histogram
    axs[row, col].set_ylabel('Frequency', fontsize=12, color='black')
    axs[row, col].set_title(variable, fontsize=14, color='black')

# Remove any extra subplots
if num_variables < len(axs.flat):
    for i in range(num_variables, len(axs.flat)):
        fig.delaxes(axs.flat[i])

# Adjust the spacing between subplots
fig.tight_layout()

# Show the plot
plt.show()
```
