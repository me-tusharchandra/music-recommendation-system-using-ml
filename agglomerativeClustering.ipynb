```python
#@title Agglomerative Clutering

import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load the data from tsne1.2.csv
data = pd.read_csv('tsne1.2.csv', nrows = 30000)

# Perform agglomerative clustering
num_clusters = 20
agg_cluster = AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward')
cluster_labels = agg_cluster.fit_predict(data)

# Add cluster labels to the original data
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = cluster_labels

# Display the cluster distribution
cluster_distribution = data_with_clusters['Cluster'].value_counts().sort_index()
print("Cluster Distribution:")
print(cluster_distribution)

# Visualization (scatter plot)
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=cluster_labels, alpha=0.7)
plt.title('Agglomerative Clustering (20 Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

#Scaling the data
df5 = pd.read_csv('tracks_features.csv', nrows = 30000)
df6 = df5.copy()

df5= df5.drop(['name', 'album', 'artists', 'artist_ids', 'id', 'album_id', 'release_date'], axis=1)

df5.head()
```

```python
scaler = MinMaxScaler()

# Normalize all columns of the dataframe
scaled_features = pd.DataFrame(scaler.fit_transform(df5), columns=df5.columns)
scaled_features

agg_cluster = AgglomerativeClustering(n_clusters=20, affinity='euclidean', linkage='ward')
df6['AGG_Labels'] = agg_cluster.fit_predict(scaled_features)
```

```python
df6.head()
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def recommend(data, given_song_features, N=10):
    # Same Cluster Recommendation:
    same_cluster_songs = data[data['AGG_Labels'] == given_song_features['AGG_Labels']]

    # If there are enough songs in the same cluster, just return the top N from them
    if len(same_cluster_songs) >= N:
        return same_cluster_songs['name'].head(N).tolist()

    # Calculate cosine similarity
    features = data.drop(columns=['name', 'album', 'artists', 'AGG_Labels'])  # Drop non-numeric columns
    given_song_features = given_song_features.drop(['name', 'album', 'artists', 'AGG_Labels'])
    cosine_similarities = cosine_similarity([given_song_features], features)[0]

    # Get indices of songs with high cosine similarity
    similar_songs_indices = np.argsort(cosine_similarities)[::-1][:N]

    # Get the recommended songs
    recommended_songs = data.loc[similar_songs_indices, 'name'].tolist()

    return recommended_songs

# Example usage
given_song_features = {
    'name': 'Testify',
    'album': 'The Battle Of Los Angeles',
    'artists': ['Rage Against The Machine'],
    'track_number': 1,
    'danceability': 0.47,
    'energy': 0.978,
    'key': 7,
    'loudness': -5.399,
    'speechiness': 0.0727,
    'acousticness': 0.0261,
    'instrumentalness': 0.000011,
    'liveness': 0.356,
    'valence': 0.503,
    'tempo': 117.906,
    'duration_ms': 210133.0,
    'year': 1999.0,
    'kmeans_labels': 0,
    'GMM_labels': 1,
    'AGG_Labels': 2  # Replace with the actual Agglomerative Clustering label
}

recommended_songs = recommend(df6, given_song_features, N=10)
print(recommended_songs)
```
